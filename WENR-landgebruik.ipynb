```python
import pandas as pd
```

```python
# Load the CSV file
file_path = 'WENR-landgebruik.csv'
data = pd.read_csv(file_path, delimiter=';', header=4)
data.columns = ['LGN-Klassen', 'LGN2018', 'LGN2021', "Verschil"]
data.head()

data
```

```python
# Keep only the columns LGN-Klassen	and LGN2021
data = data[['LGN-Klassen', 'LGN2021']]

# Remove rows that are 0
data = data[data['LGN2021'] != 0]

# Drop the row for Zout water
data = data[data['LGN-Klassen'] != 'Zout water']

data
```

```python
# Sum all the areas
total_area = data['LGN2021'].sum()

print(f'Total area: {total_area} ha')

# Calculate the percentage of each class
data['Percentage'] = round(data['LGN2021'] / total_area, 4)

data
```

```python
# Sanity check, the sum of all percentages should be 1
print(data['Percentage'].sum())
```
