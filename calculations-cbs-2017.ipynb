```python
import pandas as pd
```

### Load the CSV file

```python
# Load the CSV file
file_path = 'Bodemgebruik__uitgebreide_gebruiksvorm__per_gemeente_31052024_155011.csv'
data = pd.read_csv(file_path, delimiter=';', header=4)
data.columns = ['Gebruiksvorm', 'Eenheid', 'Oppervlakte']
data.head()

# remove last 6 row, we're not interested in buitenwater
data = data[:-6]

data
```

```python
# Filter the data, show only the rows that contain |Totaal
data_totalen = data[data['Gebruiksvorm'].str.contains('Totaal')]
data_totalen.head()

# List the table
data_totalen

# Calculate the total area of the Netherlands
total_area = data_totalen['Oppervlakte'].sum()
total_area
```

```python

# Calculate the percentage of the land area for each category
data['Percentage'] = round(data['Oppervlakte'] / total_area, 4)
data
```

```python
# List with only the rows that contain "|Totaal"

data = data[data['Gebruiksvorm'].str.contains('Totaal')]

data
```

```python
# Sum percentages to make sure they add up to 1
data['Percentage'].sum()
```
